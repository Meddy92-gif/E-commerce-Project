Creating SQLite Database


```python
db_conn = sqlite3.connect("superstore.db")
```

```python
cursor = db_conn.cursor()
```

```python
customers.head()
```

```python
cursor.execute('DROP TABLE Customer;')
```

Creating each table connected to SQLite

```python
#Customers Table
cursor.execute('''
    CREATE TABLE Customer (
        CustomerID TEXT NOT NULL,
        CustomerName TEXT,
        PostalCode INTEGER,
        Region TEXT,
        City TEXT,
        State TEXT,
        Segment TEXT,
        PRIMARY KEY(CustomerID));
''')
db_conn.commit()
```

```python
#cursor.execute('''
    #ALTER TABLE Customer
    #ADD COLUMN Segment TEXT;
#''')
#db_conn.commit()
```

```python
customers.to_sql('Customer', db_conn, if_exists='append', index=False)
```

```python
customers.head()
```

```python
#Products Table
cursor.execute('''
    CREATE TABLE Products (
        ProductID TEXT NOT NULL,
        ProductName TEXT,
        Category TEXT,
        SubCategory TEXT,
        PRIMARY KEY(ProductID)); 
''')
db_conn.commit()
```

```python
products.to_sql('Products', db_conn, if_exists='append', index=False)
```

```python
#cursor.execute('DROP TABLE Orders;')
```

```python
#table orders
cursor.execute("""
    CREATE TABLE Orders (
        OrderID TEXT NOT NULL,
        ProductID TEXT,
        CustomerID TEXT,
        OrderDate DATE,
        ShipDate DATE,
        ShipMode TEXT,
        PRIMARY KEY(OrderID),
        FOREIGN KEY(ProductID) REFERENCES Products(ProductID),
        FOREIGN KEY(CustomerID) REFERENCES Customer(CustomerID));
    """)
db_conn.commit()
```

```python
orders.head()
```

```python
orders.to_sql('Orders', db_conn, if_exists='append', index=False)
```

```python
#SalesInfo Table
cursor.execute('''
    CREATE TABLE Salesinfo (
        SalesID INTEGER,
        Sales REAL,
        CustomerID TEXT,
        OrderID TEXT,
        Quantity INTEGER,
        Discount REAL,
        Profit REAL,
        PRIMARY KEY(SalesID),
        FOREIGN KEY(OrderID) REFERENCES Orders(OrderID),
        FOREIGN KEY(CustomerID) REFERENCES Customer(CustomerID));
''')
db_conn.commit()
```

```python
salesinfo.to_sql('Salesinfo', db_conn, if_exists='append', index=False)
```
